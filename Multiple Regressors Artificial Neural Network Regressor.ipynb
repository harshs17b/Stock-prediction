```python
stock_symbol = "AAPL"
start_date = "2023-08-01"
end_date = "2023-09-01"

# Load the stock price data from Yahoo Finance
df = yf.download(stock_symbol, start=start_date, end=end_date)
print(df.shape)
```

```python
df['Day'] = b
```

```python
features = df.drop(columns=["Date", "Close"])
target = df.Close
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor, AdaBoostRegressor
from xgboost import XGBRegressor
from catboost import CatBoostRegressor
from lightgbm import LGBMRegressor
from sklearn.linear_model import SGDRegressor
from sklearn.pipeline import Pipeline
from sklearn.feature_selection import SelectFromModel
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error

```

```python
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.25, random_state=42)
```

```python
names = ["Linear Regression", "Decision Tree Regressor", "Random Forest Regressor", "GradientBoosting Regressor", "AdaBoost Regressor"
        , "XGBoost Regressor", "CatBoost Regressor", "LightGBM Regressor", "SGD Regressor "]
```

```python
Regressor = [
    LinearRegression(),
    DecisionTreeRegressor(),
    RandomForestRegressor(),
    GradientBoostingRegressor(),
    AdaBoostRegressor(),
    XGBRegressor(),
    CatBoostRegressor(),
    LGBMRegressor(),
    SGDRegressor()
]

```

```python
scores = []
rmse = []
```

```python
for reg in Regressor:
    pipe = Pipeline([
        ("scaling", StandardScaler() ),
        ("SelectFromModel", SelectFromModel(estimator=RandomForestRegressor(n_estimators=100))),
        ("Regressor", reg)
    ])
    
    pipe.fit(X_train, y_train)
    
    y_pred = pipe.predict(X_test)
    
    print(reg)
    scores.append(pipe.score(X_test, y_test))
    print("R2 score is")
    print(pipe.score(X_test, y_test))
    
    rmse.append(mean_squared_error(y_test, y_pred))
    print("mean squared error is")
    print(mean_squared_error(y_test, y_pred, squared=False))
    
    
    
    
```

```python
dfResult = pd.DataFrame(zip(names, scores, rmse), columns=["names", "scores", "rmse"])
```

```python
dfResult
```

```python
dfResult.sort_values(by=['scores', 'rmse'], ascending=[False,True])
```

### Selecting the best pipe

```python
pipe = Pipeline([
    ("scaling", StandardScaler()),
    ("SelectBest", SelectFromModel(estimator=RandomForestRegressor(n_estimators=100))),
    ("model", SGDRegressor())
])

pipe.fit(X_train, y_train)

y_pred = pipe.predict(X_test)

pipe.score(X_test, y_test)

```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.scatter(y_pred, y_test, alpha=.6, color='r') 
plt.xlabel('Predicted Price')
plt.ylabel('Actual Price')
plt.title('Linear Regression Model')
```
